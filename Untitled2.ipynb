```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_excel("recommendation_results_2.xlsx")
data=data[['Recommended Recipe','Healthiness Score','protein','calories','fat','sodium','Diabetes']]
df = pd.DataFrame(data)

# Plotting the heatmap
sns.set(style="whitegrid")
plt.figure(figsize=(20, 8))
columns_to_visualize = ['protein', 'fat', 'sodium', 'calories', 'Healthiness Score']

# Creating the heatmap
heatmap_data = df[columns_to_visualize]
heatmap = sns.heatmap(heatmap_data.corr(), annot=True, cmap='coolwarm', linewidths=.5, fmt=".2f")

plt.title('Correlation Heatmap with Healthiness Score')
plt.show()
```

```python
df_no_duplicity = df.drop_duplicates()
df_no_duplicity = df_no_duplicity[(df_no_duplicity["calories"] < df_no_duplicity.calories.quantile(0.95)) & (df_no_duplicity["calories"] >0)]
def plot_hist(axes, x, label, xlabel):
  axes.hist(x, histtype="step", label=label)
  axes.set_xlabel(xlabel)
  axes.legend()
fig, ax = plt.subplots(2,2,figsize=(16, 8))
plot_hist(ax[0,0], df[df.Diabetes == 0]["calories"], "not diabetes", "calories")
plot_hist(ax[0,0], df[df.Diabetes == 1]["calories"], "diabetes", "calories")
plot_hist(ax[0,1], df[df.Diabetes == 0]["protein"], "not diabetes", "protein")
plot_hist(ax[0,1], df[df.Diabetes == 1]["protein"], "diabetes", "protein")
plot_hist(ax[1,0], df[df.Diabetes == 0]["fat"], "not diabetes", "fat")
plot_hist(ax[1,0], df[df.Diabetes == 1]["fat"], "diabetes", "fat")
plot_hist(ax[1,1], df[df.Diabetes == 0]["sodium"], "not diabetes", "sodium")
plot_hist(ax[1,1], df[df.Diabetes == 1]["sodium"], "diabetes", "sodium")
```
